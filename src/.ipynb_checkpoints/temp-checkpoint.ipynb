## 处理现有信息


!pip install cpca

```python
residences=list(df_new['现居地'])
```

```python
import cpca
location_str = residences
df = cpca.transform(location_str, pos_sensitive=True)
df
```

```python
for col in ['省', '市', '区']:
    df_new[f'现居地_{col}']=list(df[col])
```

```python
for ind,row in df_new.loc[df_new['现居地_省'].isnull(),].iterrows():
    location_str = [row['出生地']]
    df = cpca.transform(location_str, pos_sensitive=True)
    df_new.loc[ind, f'现居地_省'] = df['省'].values[0]
    df_new.loc[ind, f'现居地_市'] = df['市'].values[0]
    df_new.loc[ind, f'现居地_区'] = df['区'].values[0]
```

```python
df_new = df_new.loc[~df_new['现居地_省'].isnull(),]
```

```python
df_new.columns
```

```python
df_new=df_new.drop(df_new.loc[df_new['现居地_省']=='台湾省',].index,)
```

```python
df_new.to_csv(Path.cwd().parent/'Data/crime_data_reasons_selected_no_na_location_identified_no_na.csv',index=False)
```

```python
for ind,row in df.iterrows():
    control = False 
    reason = row['案由']
    for key in replace_dict.keys():
        if key in reason:
            replaced_reason = reason.replace(key,replace_dict[key])
            df.loc[ind,'案由_替换']=replaced_reason.replace('罪','')
            control = True
    if control:
        reason = replaced_reason
    if '诈骗' in reason:
        reasons = reason.split('、')
        for r in reasons:
            if '诈骗' in r:
                reasons.remove(r)
                reasons.append('Fraud')
        replaced_reason = '、'.join(reasons)
        df.loc[ind,'案由_替换']=replaced_reason
    
```
