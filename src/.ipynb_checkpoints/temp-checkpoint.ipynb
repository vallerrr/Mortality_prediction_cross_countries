```python
import pandas as pd 
import numpy as np
from pathlib import Path 
import matplotlib
pd.set_option('display.max_rows', 1000)
Path.cwd()
```

```python
df = pd.read_csv(Path.cwd().parent/'Data/crime_data.csv')
```

# 数据清理

```python
df.replace({'暂无数据':None,'住':None,'出生于':None,'案由':None,'其他刑事犯':None,'数据缺失':None},inplace=True)
df.dropna(subset=['案由'],inplace=True)
```

Restore data from less to more, using '案由'

```python
df_new=df.copy()
df_new.dropna(subset=['案由'],inplace=True)
df_new = df_new.loc[df_new['案由'].str.contains('诈骗'),]
```

```python
reasons = ['网络','信息','伪造、变造、买卖国家机关公文、证件、印章',
 '伪造、变造、买卖国家机关公文、证件、印章、伪造公司、企业、事业单位、人民团体印章',
 '伪造、变造、买卖国家机关公文、证件、印章罪',
 '伪造、变造、买卖身份证件罪',
 '伪造、变造居民身份证罪',
 '伪造、变造金融票证',
 '伪造、变造金融票证罪',
 '伪造公司、企业、事业单位、人民团体印章',
 '伪造公司、企业、事业单位、人民团体印章、损害商业信誉、商品声誉、敲诈勒索',
 '伪造公司、企业、事业单位、人民团体印章罪',
 '伪造货币罪',
 '使用虚假身份证件、盗用身份证件',
 '使用虚假身份证件、盗用身份证件罪','妨害信用卡管理',
 '妨害信用卡管理、掩饰、隐瞒犯罪所得、犯罪所得收益',
 '妨害信用卡管理、盗窃',
 '妨害信用卡管理罪',
    '掩饰、隐瞒犯罪所得、犯罪所得收益',
 '掩饰、隐瞒犯罪所得、犯罪所得收益、妨害信用卡管理',
 '掩饰、隐瞒犯罪所得、犯罪所得收益、妨害信用卡管理、窝藏、包庇',
 '掩饰、隐瞒犯罪所得、犯罪所得收益罪',
'破坏广播电视设施、公用电信设施罪',
 '窝藏、包庇',
 '窝藏、包庇罪',
 '窝藏、转移、收购、销售赃物罪','隐瞒、掩饰犯罪所得、犯罪所得收益',
 '隐瞒、掩饰犯罪所得、犯罪所得收益罪','非法吸收公众存款',
 '非法吸收公众存款、开设赌场',
 '非法吸收公众存款罪','非法经营','传授犯罪方法']
```

```python
for reason in reasons:
    temp=df.loc[df['案由'].str.contains(reason),]
    df_new= pd.concat([temp,df_new],axis=0)
    df_new.drop_duplicates(inplace=True)
```

```python
df_new=df_new.dropna()
df_new.to_csv(Path.cwd().parent/'Data/crime_data_reasons_selected_no_na.csv')
```

# 获取市县数据

## 处理现有信息


!pip install cpca

```python
residences=list(df_new['现居地'])
```

```python
import cpca
location_str = residences
df = cpca.transform(location_str, pos_sensitive=True)
df
```

```python
for col in ['省', '市', '区']:
    df_new[f'现居地_{col}']=list(df[col])
```

```python
for ind,row in df_new.loc[df_new['现居地_省'].isnull(),].iterrows():
    location_str = [row['出生地']]
    df = cpca.transform(location_str, pos_sensitive=True)
    df_new.loc[ind, f'现居地_省'] = df['省'].values[0]
    df_new.loc[ind, f'现居地_市'] = df['市'].values[0]
    df_new.loc[ind, f'现居地_区'] = df['区'].values[0]

```

```python
df_new = df_new.loc[~df_new['现居地_省'].isnull(),]
```

```python
df_new.to_csv(Path.cwd().parent/'Data/crime_data_reasons_selected_no_na_location_identified_no_na.csv',index=False)
```
